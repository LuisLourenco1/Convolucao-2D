## Implementando o algoritmo da convolução 2D

### Importando Bibliotecas

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Importando imagem

```python
img = cv2.imread('img.bmp')  # armazena a imagem na variável

img = cv2.cvtColor(src=img, code=cv2.COLOR_BGR2GRAY)  # converte a imagem para escalas de cinza (aparentemente a imagem ainda estava como RGB)

plt.imshow(img, cmap='gray')  # plota a imagem
```

### Criando a Convolução 2D

```python
def convolucao2D(imagem, filtro, padding=0, strides=1):
    # espelhamento
    filtro = np.flipud(np.fliplr(filtro))

    # coletando formas do tipo + imagem + preenchimento
    xFiltroShape = filtro.shape[0]
    yFiltroShape = filtro.shape[1]
    xImgShape = imagem.shape[0]
    yImgShape = imagem.shape[1]

    # forma da saída (output) da convolução
    xOutput = int(((xImgShape - xFiltroShape + 2 * padding) / strides) + 1)
    yOutput = int(((yImgShape - yFiltroShape + 2 * padding) / strides) + 1)
    output = np.zeros((xOutput, yOutput))

    # aplica preenchimento igual em todos os lados da imagem
    if padding != 0:
        imagemPreenchida = np.zeros((imagem.shape[0] + padding*2, imagem.shape[1] + padding*2))
        imagemPreenchida[int(padding):int(-1 * padding), int(padding):int(-1 * padding)] = imagem
        print(imagemPreenchida)
    else:
        imagemPreenchida = imagem

    # itera através da imagem
    for y in range(imagem.shape[1]):
        # sai da convolução
        if y > imagem.shape[1] - yFiltroShape:
            break
        # só realiza a convolução se y tiver diminuído através dos Strides especificados
        if y % strides == 0:
            for x in range(imagem.shape[0]):
                # vai para a próxima linha quando kernel está fora dos limites
                if x > imagem.shape[0] - xFiltroShape:
                    break
                try:
                    # só realiza a convolução se x passou pelos Strides especificados
                    if x % strides == 0:
                        output[x, y] = (filtro * imagemPreenchida[x: x + xFiltroShape, y: y + yFiltroShape]).sum()
                except:
                    break

    return output
```

### Aplicando o Filtro da Média 5x5

```python
# filtro da média 5x5
FiltroMedia = np.array([[1/25, 1/25, 1/25, 1/25, 1/25], [1/25, 1/25, 1/25, 1/25, 1/25], [1/25, 1/25, 1/25, 1/25, 1/25], [1/25, 1/25, 1/25, 1/25, 1/25], [1/25, 1/25, 1/25, 1/25, 1/25]])

outputFM = convolucao2D(img, FiltroMedia)

plt.imshow(outputFM, cmap='gray')
```

### Aplicando o Filtro Gaussiano 5x5

```python
# filtro gaussiano 5x5
FiltroGaussiano = np.array([[1, 4, 6, 4, 1], [4, 16, 24, 16, 4], [6, 24, 36, 24, 6], [4, 16, 24, 16, 4], [1, 4, 6, 4, 1]])

outputFG = convolucao2D(img, FiltroGaussiano)

plt.imshow(outputFG, cmap='gray')
```

### Aplicando o Filtro Laplaciano 3x3

```python
# filtro laplaciano 3x3
FiltroLaplaciano = np.array([[0, 1, 0], [1, -4, 1], [0, 1, 0]])

outputFL = convolucao2D(img, FiltroLaplaciano)

plt.imshow(outputFL, cmap='gray')
```

### Filtro Laplaciano da Gaussiana 7x7

```python
# filtro laplaciano da gaussiana 7x7
FiltroLaplacianoDaGaussiana = np.array([[0, 0, 1, 1, 1, 0, 0], [0, 1, 3, 3, 3, 1, 0], [1, 3, 0, -7, 0, 3, 1], [1, 3, -7, -24, -7, 3, 1], [1, 3, 0, -7, 0, 3, 1], [0, 1, 3, 3, 3, 1, 0], [0, 0, 1, 1, 1, 0, 0]])

outputFLG = convolucao2D(img, FiltroLaplacianoDaGaussiana)

plt.imshow(outputFLG, cmap='gray')
```
